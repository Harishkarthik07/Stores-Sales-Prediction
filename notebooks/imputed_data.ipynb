```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
```

```python
train_df=pd.read_csv('Train.csv')
```

```python
train_df['Item_Weight'] = train_df['Item_Weight'].fillna(train_df['Item_Weight'].mean())
```

```python
train_df['Outlet_Size'] = train_df['Outlet_Size'].fillna('Medium')
```

```python
train_df['Item_Visibility'] = train_df['Item_Visibility'].replace(0, train_df['Item_Visibility'].mean())
```

```python

label_encoders = {}
for col in ['Item_Fat_Content', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type']:
    le = LabelEncoder()
    train_df[col] = le.fit_transform(train_df[col])
    label_encoders[col] = le
```

```python
train_df['Item_Fat_Content'] = train_df['Item_Fat_Content'].replace({'low fat': 'Low Fat','LF': 'Low Fat','reg': 'Regular'})
```
